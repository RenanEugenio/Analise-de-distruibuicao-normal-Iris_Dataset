```python
from scipy import stats
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
iris = load_iris()
```

Verificacao se as variaes do datasets iris, estao normalmente distribuidos

```python
# Analisando o histograma

plt.figure(1)
plt.subplot(2, 2, 1)
srn.histplot(iris.data[:, 0], bins=20, color='green',).set_title('sepal length')
plt.subplot(2, 2, 2)
srn.histplot(iris.data[:, 1], bins=20, color='red').set_title('sepal width')
plt.subplot(2, 2, 3)
srn.histplot(iris.data[:, 2], bins=20, color='yellow').set_title('petal length')
plt.subplot(2, 2, 4)
srn.histplot(iris.data[:, 3], bins=20).set_title('petal width')
plt.tight_layout()
plt.show()
```

```python
list = ['sepal length', 'sepal width', 'petal length', 'petal width']
```

```python
# Diagrama de Normalidade
# No entanto podemos utilizar outro grafico mais apropriado para analisar
# se as variáveis têm um comportamento de uma distribuição normal

for i in range(len(list)):
    fig, ax = plt.subplots()
    stats.probplot(iris.data[:, i], fit=True, plot=ax)
    ax.set_title(list[i])
    plt.show()
```

```python
# Por fim, e executado o teste de shapiro para de fato analisar se 
# as variaveis comportam-se com uma distribuicao normal

for i in range(len(list)):
    spr = stats.shapiro(iris.data[:, i])
    print(f'{list[i]}: {spr}')
```

Concluímos que, apenas a variável 'sepal width' apresenta uma distribuição normal. No entanto,
'sepal length', apresentou tendência de normalidade nas representações gráficas, mas seu p value
no teste de shapiro foi menor que 0.05, desta forma, a hipótese foi aceita.
